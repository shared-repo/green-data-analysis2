```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
from nltk import WhitespaceTokenizer
from collections import Counter
```

```python
with open("data-files/constitution-en.txt", "rt") as f:
    constitution = f.read()
constitution    
```

```python
print( len(constitution) ) # 문자 갯수
print( constitution[:40] )
```

```python
len(STOPWORDS), type(STOPWORDS)
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS)
result = wordcloud.generate(constitution)
```

```python
print( result )
print( result.words_ )
print( len(result.words_) )
```

```python
plt.figure(figsize=(10, 7))
plt.imshow(result)
plt.axis('off')
plt.show()
```

```python
with open("data-files/constitution-ko.txt", "rt") as f:
    constitution = f.read()
```

```python
len(constitution), constitution[:40]
```

```python
tokenizer = WhitespaceTokenizer()
constitution_tokens = tokenizer.tokenize(constitution)
```

```python
constitution_tokens[100:200]
```

```python
for w in ['①', '②', ".", ",", ";", "\"", "\'", "<", ">", "(", ")", '수', '또는', '의하여', '있다', '한다', '그']:
    constitution_tokens = list( map(lambda word : word.replace(w, ""), constitution_tokens) )
```

```python
constitution_tokens[100:110]
```

```python
constitution_tokens = list( filter(lambda word: len(word) > 0, constitution_tokens) )
# constitution_tokens = list(constitution_tokens)
```

```python
len(constitution_tokens)
```

```python
counter = Counter(constitution_tokens)
```

```python
print( counter.most_common(100)[:10] )
most_common_tokens = dict(counter.most_common(200))
```

```python
most_common_tokens
```

```python
wordcloud = WordCloud().generate_from_frequencies(most_common_tokens) # 단언 빈도수 dict로 wordcloud 만들기
```

```python
wordcloud.words_
```

```python
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
