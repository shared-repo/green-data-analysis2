```python
from urllib import request # web request handling tool ( 내장 라이브러리 )
import requests # web request handling tool ( 추가 설치 필요 )
```

```python
response = request.urlopen("http://developers.google.com")
bhtml = response.read()
chtml = bhtml.decode("utf-8")
chtml
```

```python
response = requests.get("http://developers.apple.com")
html = response.content.decode("utf-8")
html = html.replace("\n", "").replace("\t", "")
html
```
