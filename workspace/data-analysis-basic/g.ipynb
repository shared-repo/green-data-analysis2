```python
# 3개의 데이터 테이블에서 집계행 제거
# 기타, 교포 행 제거 (선택적)
# 목적별(연령별, 성별) 여행객 수 조회
# 국가별 여행객 수 조회
# 대륙별 여행객 수 조회
# 년도별 여행객 수 조회
```

```python
# 집계행 제거 함수
def remove_sum_rows(df, col):
    sum_mask = df["국가명"].str.contains('연도')    
    sum_mask.fillna(True, inplace=True)
    sum_mask2 = df[col].str.contains("전체")
    return df[~(sum_mask|sum_mask2)]
```

```python
#함계 행 제거 처리
tourist_by_age2 = remove_sum_rows(tourist_by_age, "연령")
tourist_by_purpose2 = remove_sum_rows(tourist_by_purpose, "목적구분")
tourist_by_gender2 = remove_sum_rows(tourist_by_gender, "성별")
```

```python
tourist_by_purpose2['목적구분'].value_counts().index
```

```python
# 목적별 여행객 수 조회 1
tourist_by_purpose2["목적구분"].value_counts()
for purpose in tourist_by_purpose2['목적구분'].value_counts().index:
    purpose_mask = tourist_by_purpose2["목적구분"] == purpose
    sum_by_purpose = tourist_by_purpose2[purpose_mask]['인원'].sum()
    print(purpose, ":", sum_by_purpose)
```

```python
# 목적별 여행객 수 조회 2 ( 다른 데이터프레임에 적용할 수 있도록 일반화 : 함수 만들기 )

def calculate_sum_by_col(df, col):
    result = {}
    for purpose in df[col].value_counts().index:
        purpose_mask = df[col] == purpose
        sum_by_purpose = df[purpose_mask]['인원'].sum()
        result.update({purpose : sum_by_purpose})
    return result

calculate_sum_by_col(tourist_by_age2, "연령")
```
