```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256)
C = np.cos(X)
plt.plot(X, C)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.show()
```

```python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--')
plt.plot(t, 0.5*t**2, 'bs:')
plt.plot(t, 0.2*t**3, 'g^-')
plt.show()
```
