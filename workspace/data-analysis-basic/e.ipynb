```python
# 무임 승객 인원이 가장 많은 역 10개
# 무임 승객 비율이 가장 많은 역 10개
# 1. 무임승하차 컬럼 만들기 : 무임승차 + 무임하차
```

```python
# 유임승하차, 무임승하차 컬럼 만들기
subway_fee["무임승하차"] = \
    subway_fee['무임승차'].map(lambda v: int(v.replace(",", ""))) + \
    subway_fee['무임하차'].map(lambda v: int(v.replace(",", "")))
subway_fee["유임승하차"] = \
    subway_fee['유임승차'].map(lambda v: int(v.replace(",", ""))) + \
    subway_fee['유임하차'].map(lambda v: int(v.replace(",", "")))
subway_fee["유임승하차"].dtype
```

```python
# 무임승하차비율 컬럼 만들기
subway_fee["무임승하차비율"] = \
    subway_fee['무임승하차'] / (subway_fee["유임승하차"] + subway_fee['무임승하차'])
```

```python
# 무임승하차 컬럼 기준 정렬
sorted_values = subway_fee["무임승하차"].sort_values(ascending=False)
type(sorted_values)
sorted_values
```

```python
# 정렬 -> 상위 10개 조회 1
top10_mask = subway_fee['무임승하차'].map(lambda v: v in sorted_values[:10].values)
subway_fee[top10_mask]
```
